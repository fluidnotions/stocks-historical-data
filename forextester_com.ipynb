```python
csvFileName = 'EURUSD'
```

```python
import pandas as pd
#https://forextester.com/data/
def forextester_free_data_transform(data):
    transformed = data.copy()
    # print(f'{transformed.shape} rows before filtering <TIME> length = 6')
    rows = transformed.shape[0]
    # Filter out rows where the length of the <TIME> column is not equal to 6
    transformed = transformed[transformed['<TIME>'].astype(str).str.len() == 6]
    # print(f'{transformed.shape} rows after filtering <TIME> length = 6')
    rowsAfter = transformed.shape[0]
    print(f'{rows - rowsAfter} rows filtered out, due to <TIME> length not equal to 6')
    time_format = '%H%M%S'
    date_format = '%Y%m%d'
    date_time_format = f'{date_format} {time_format}'
    # Convert the columns to datetime objects
    transformed['Date'] = pd.to_datetime(transformed['<DTYYYYMMDD>'].astype(str) + " " +transformed['<TIME>'].astype(str), format=date_time_format)
    # Drop the original <TIME> column
    transformed = transformed.drop(columns=['<TIME>', '<DTYYYYMMDD>', '<TICKER>'])
    # Set the Date column as the index
    transformed.set_index('Date', inplace=True)
    # Define the new column names
    new_columns = {'<OPEN>': 'Open', '<HIGH>': 'High', '<LOW>': 'Low', '<CLOSE>': 'Close', '<VOL>': 'Volume'}
    # Rename the columns
    transformed.rename(columns=new_columns, inplace=True)
    return transformed
```

```python
data = pd.read_csv(f'./{csvFileName}.csv')
data.tail()
```

```python
cleaned = forextester_free_data_transform(data)
cleaned
```

```python
import os

outFile = f'./{csvFileName}-cleaned.csv'
if os.path.exists(outFile):
    os.remove(outFile)
```

```python

cleaned.to_csv(outFile)
```
